```python
%pylab inline
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test.jpg')
```

```python
Gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(Gray)
```

```python
plt.show()
```

```python
_, equ = cv2.threshold(Gray,200,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(equ)
```

```python
plt.show("gray",equ)
```

```python
canny = cv2.Canny(equ, 100, 200)
```

```python
plt.subplot(122),plt.imshow(canny,cmap = 'gray')
```

```python
plt.show()
```
